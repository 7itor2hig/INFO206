```python
file = "doi.txt"
num_words = 0
num_char = 0

#open file, read through it and count the number of lines, words, and characters
with open(file, "r") as text: #opens file and reads it as text
    
    #converts open file to a single string
    text = text.read()
    
    #splits the string by new line character and counts the number of new lines
    lines = text.split("\n")
    num_lines = len(lines)-1
    
    #replaces non-word characters in the full string with blank spaces
    words = text.replace(",", " ")
    words = words.replace("\n", " ")
    
    #splits the text by spaces and adds them to a list
    words = words.split(" ")
    
    #counts the words in the list that are not spaces
    for i in words:
        if i != "":
            num_words += 1
    
    #replaces only the new line character in the text
    char = text.replace("\n", " ")
    
    #counts all characters in the text that are not spaces 
    for i in char:
        if i != " ":
            num_char +=1
            
#prints the number of lines, words, and characters in the whole document            
print(str(num_lines) + " lines; " + str(num_words) + " words; " + str(num_char) + " characters.")
```
